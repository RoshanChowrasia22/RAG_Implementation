```python
import faiss
import numpy as np
```

### IVF using PQ

```python
n=1000
d=768
embeddings=np.random.rand(n,d).astype('float32')
nlist=10 # IVF clusters (K)
m=8        # PQ sub-vectors
bits=8     # 256 codes
nprobe=5  # Clusters to probe

# IVF quantizer (for centriods)
quantizer=faiss.IndexFlatL2(d)
index=faiss.IndexIVFPQ(quantizer,d,nlist,m,bits)

index.train(embeddings)
index.add(embeddings)
```

```python
# Query
query=np.random.rand(1,d).astype('float32')
k=2
# Set nprobe as an attribute (not passed to search)
index.nprobe = nprobe
D,I=index.search(query,k)
print(f"Top-{k} indices: {I[0]}")
print(D)
```
